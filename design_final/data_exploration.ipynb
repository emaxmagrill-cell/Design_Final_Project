# Team Statistics Dataframe Exploration

```python
team_stats = nfl.load_schedules(2024).to_pandas()
team_stats = pd.DataFrame(team_stats)
team_stats
```

```python
team_stats.columns
```

```python
df = team_stats[team_stats['game_type'] == 'REG']
teams = pd.unique(df[['home_team', 'away_team']].values.ravel())
win_counts = {team: 0 for team in teams}
for _, row in df.iterrows():
    if row['home_score'] > row['away_score']:
        win_counts[row['home_team']] += 1
    elif row['away_score'] > row['home_score']:
        win_counts[row['away_team']] += 1
wins_df = (
    pd.DataFrame.from_dict(win_counts, orient='index', columns=['wins'])
      .sort_values('wins', ascending=False)
)
wins_df
```

# Second Figure Attempt

```python
injury_counts = injury_data['team'].value_counts()
combined = pd.DataFrame({
    'injuries': injury_counts,
    'wins': wins_df['wins']
}).fillna(0)
combined = combined.sort_values("wins", ascending=False)
teams = combined.index
inj = combined['injuries']
wins = combined['wins']
mean_injuries = inj.mean()
x = np.arange(len(teams))
width = 0.40
plt.figure(figsize=(16,6))
plt.bar(x - width/2, inj, width, label='Injuries')
plt.bar(x + width/2, wins, width, label='Wins')
plt.axhline(mean_injuries, color='red', linestyle='--', linewidth=2, label=f"Mean Injuries ({mean_injuries:.1f})")
plt.ylim(0, max(inj.max(), wins.max(), mean_injuries) * 1.15)
plt.xticks(x, teams, rotation=45)
plt.ylabel("Count")
plt.title("Injuries vs Wins per Team (Grouped, Sorted by Wins)")
plt.legend()
plt.tight_layout()
plt.show()
```

# Five Figure

```python
def get_playoff_teams():
    playoff = {}
    for year in range(2013, 2025):
        depth = nfl.load_depth_charts(year).to_pandas()
        playoff_teams = (
            depth[depth["game_type"] != "REG"]["club_code"].dropna().unique().tolist()
        )
        playoff[year] = sorted(playoff_teams)
    return playoff

def get_starting_qbs_for_year(year):
    snaps = nfl.load_snap_counts(year).to_pandas()
    snaps_qb = snaps[snaps["position"] == "QB"]
    starting_qbs = (
        snaps_qb.groupby("team")["player"].agg(lambda x: x.value_counts().idxmax()).to_dict()
    )
    return starting_qbs

def get_weeks_played_for_year(year, starting_qbs):
    snaps = nfl.load_snap_counts(year).to_pandas()
    snaps = snaps[snaps["game_type"] == "REG"]
    snaps = snaps[snaps["player"].isin(starting_qbs.values())]
    weeks_played = snaps.value_counts("team")
    return weeks_played.to_dict()

def compute_percentages(playoff_teams, qb_games):
    percentages = {}
    for year in range(2013, 2025):
        teams = playoff_teams.get(year, [])
        if not teams:
            percentages[year] = 0
            continue
        total = len(teams)
        good = sum(
            qb_games.get(year, {}).get(team, 0) >= 16
            for team in teams
        )
        percentages[year] = (good / total) * 100
    return percentages

def plot_percentages(percentages):
    years = list(percentages.keys())
    vals = [percentages[y] for y in years]

    plt.figure(figsize=(12, 6))
    plt.bar(years, vals, color="#013369")

    plt.ylabel("Percentage of Playoff Teams with QB playing 16 games (%)", fontsize=14)
    plt.xlabel("Season", fontsize=18)
    plt.title("Percentage of Playoff Teams Whose Starting QB Played ≥ 16 Games", fontsize=20)
    plt.ylim(0, 100)
    plt.xticks(years, rotation=60, fontsize=17)
    plt.yticks(fontsize=17)

    plt.tight_layout()
    plt.show()

playoff_teams = get_playoff_teams()
qb_games = {}
for year in range(2013, 2025):
    starting_qbs = get_starting_qbs_for_year(year)
    weeks = get_weeks_played_for_year(year, starting_qbs)
    qb_games[year] = weeks

percentages = compute_percentages(playoff_teams, qb_games)
plot_percentages(percentages)
```

# Next Gen Stats exploration and graph experimentation

```python
next_gen = nfl.load_team_stats(2024).to_pandas()
next_gen = next_gen[next_gen['season_type'] == 'REG']
dictionary = next_gen.groupby('team')['passing_yards'].sum().to_dict()
dictionary
```

```python
snaps = nfl.load_snap_counts(2024).to_pandas()
qb_snaps = snaps[snaps['player'].isin(starting_qbs_2024)]
qb_snaps = qb_snaps[qb_snaps['game_type'] == 'REG']
weeks_starting_qb_played = qb_snaps.value_counts("team")
weeks_starting_qb_played
```

```python
passing_yard_games_played = pd.DataFrame({
    "games_played": weeks_starting_qb_played,
    "passing_yards": pd.Series(dictionary)
})
passing_yard_games_played
```

```python
from matplotlib.colors import LinearSegmentedColormap
import seaborn as sns
import matplotlib.pyplot as plt

# --- Custom gradient colormap ---
blue_red_cmap = LinearSegmentedColormap.from_list(
    "blue_red",
    ["#013369", "#D50A0A"]
)

right_side = {"PIT", "NO", "HOU", "SEA"}

# Align wins with your data
passing_yard_games_played["wins"] = wins_df["wins"].reindex(
    passing_yard_games_played.index
)

# --- Color-coded scatter with custom gradient ---
sns.scatterplot(
    data=passing_yard_games_played,
    x="games_played",
    y="passing_yards",
    hue="wins",
    palette=blue_red_cmap
)

# --- Text labels ---
for idx, row in passing_yard_games_played.iterrows():

    if idx in right_side:
        x_offset = 0.1
        y_offset = 0
    elif idx == "WAS":
        x_offset = 0.1
        y_offset = -100
    else:
        x_offset = -0.5
        y_offset = 0

    plt.text(
        row["games_played"] + x_offset,
        row["passing_yards"] + y_offset,
        idx,
        fontsize=9,
        va="center"
    )

# --- Regression line ---
sns.regplot(
    data=passing_yard_games_played,
    x="games_played",
    y="passing_yards",
    scatter=False,
    ci=None,
    line_kws={"linewidth": 2, "color": "#013369"},
    label="Correlation"
)

plt.xlabel("Games Played by Starting QB")
plt.ylabel("Total Passing Yards")
plt.title("QB Games Played vs Passing Yards (Colored by Wins)")
plt.show()
```

```python
team_stats = nfl.load_team_stats().to_pandas()
team_stats
```

```python
avg_passing_epa = (
    team_stats.groupby("team")["passing_epa"]
         .mean()
         .to_frame("avg_passing_epa")
)
avg_passing_epa
```

```python
EPA = passing_yard_games_played.join(avg_passing_epa, how="left")
EPA.to_csv("EPA.csv")
```

```python
df = EPA.copy()     # your DataFrame where index = team

# --- Create the blue → red gradient ---
blue_red_cmap = LinearSegmentedColormap.from_list(
    "blue_red_gradient",
    ["#013369", "#D50A0A"]
)

plt.figure(figsize=(12, 8))

# --- Scatter plot with gradient palette ---
sns.scatterplot(
    data=df,
    x="games_played",
    y="avg_passing_epa",
    hue=df["wins"],
    palette=blue_red_cmap,
    s=150
)

# --- Text labels ---
for team, row in df.iterrows():   # team is the index
    x = row["games_played"]
    y = row["avg_passing_epa"]

    if team in {"LAC", "CHI"}:
        x_offset, y_offset = -0.45, -0.05
    elif team in {"WAS"}:
        x_offset, y_offset = -0.45, -0.4
    else:
        x_offset, y_offset = 0.12, -0.02

    plt.text(
        x + x_offset,
        y + y_offset,
        team,
        fontsize=12,
        va="center"
    )

# --- Axis & title ---
plt.title("Games Played vs Average Passing EPA", fontsize=20)
plt.xlabel("Games Played", fontsize=20)
plt.ylabel("Average Passing EPA", fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.show()

```
